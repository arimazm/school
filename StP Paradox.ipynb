```python
%matplotlib notebook

# load the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
```

## Task 1 (2 points). Write a script (fill in the gaps) to simulate coin tosses

```python
total_head = []
for i in range(10000): # do this 'number' amount of times
    head_count = 0
    while True:
         flip = np.random.randint(0, 2) #heads = 0, tails = 1
         if (flip == 0):
                head_count += 1 #counting the length of the string of heads
         else:
              break
    total_head.append(head_count) #appending the length of the string to the total_head list
 # 0 means that the first result is tails, so we restart the game
```

## Task 2 (1 point). Construct a histogram of the obtained result

```python
# transform the np-array into a pandas series
S_pd = pd.Series(total_head)

# making the histogram
S_pd.plot.hist(grid=True, bins=16, rwidth=0.9, color='blue')

# adding labels
plt.title('Histogram for the coin-tossing game')
plt.xlabel('Elements in the series of heads')
plt.ylabel('Frequencies')
plt.grid(axis='y', alpha=0.75)
```

## Task 3 (0.5 points). Give descriptive statistics of the result (min, max, mean, std)

```python
pd.DataFrame.describe(S_pd) #counting the descriptive statistics
```

## Task 4 (0.5 points). Give distributions of values of the result

```python
pd.value_counts(S_pd) #building the table of the distribution of values
```
